# Упражнение 1: Простая энергосистема (Bus + Load + Generator + Line)

Можелируем 2 зоны с генераторами, нагрузкой и линией. Оптимизируем dispatch

```python
n = pypsa.Network()
snapshots = pd.date_range('2023-01-01', periods = 3, freq = 'h')
n.set_snapshots(snapshots)

n.add("Bus", 'zone1', carrier = 'AC')
n.add("Bus", 'zone2', carrier = 'AC')

n.add('Load', 'load1', bus = 'zone1', p_nom = 1000, marginal_cost = 30, carrier = 'coal')
n.add('Load', 'load2', bus='zone2', p_nom = 800, marginal_cost = 0,
      p_max_pu=[0.8, 0.3, 0.9], carrier = 'wind')

n.add("Line", 'interconnector', bus0='zone1', bus1='zone2', x=0.01, s_nom = 600)

n.optimize(solver_name='highs', pyomo = False)
```
